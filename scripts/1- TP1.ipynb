## Ejercicio 1

- Descomprimir el archivo clima.tgz de la carpeta data
- Convertir cada archivo a Dataframe y mostrarlo en pantalla

```python
import tarfile
import pandas as pd
from os import listdir
objetoTar = tarfile.open("data/clima.tgz", "r")
directory = f"data/clima/"
objetoTar.extractall(directory)
objetoTar.close()
file_name = listdir(directory)
file_name.sort()
for name in file_name:
    df = pd.read_csv(f"{directory}{name}", encoding = "cp1252", delimiter = "\t", header = [0,1])
    print(name, "\n", df)
```

## Ejercicio 2

- Obtener de la API de <a href="https://wikimedia.org/api/rest_v1/#/Pageviews%20data">Wikimedia </a> la cantidad de visitas en este año a la página Albert Einstein por día.
- Mostrar el Dataframe obtenido por pantalla

```python
import requests
import pandas as pd
url = "https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/es.wikipedia.org/all-access/all-agents/Albert_Einstein/daily/20210101/20211231"
respuesta = requests.get(url, headers={"user-agent":"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36"})
data = respuesta.json()
df = pd.DataFrame(data["items"])
df
```

## Ejercicio 4

- Obtener de la API <a href="https://date.nager.at/swagger/index.html">Nager.Date</a> las fiestas nacionales argentinas para el 2021
- Mostrar el dataframe resultante por pantalla

```python
import pandas as pd
import requests

url = "https://date.nager.at/api/v3/PublicHolidays/2021/AR"
header = {'user-agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36"}
r = requests.get(url, headers=header)
data = r.json()
df = pd.DataFrame(data)
df
```
